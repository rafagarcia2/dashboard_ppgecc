```python
def convert_subject_areas_to_list(subject_areas_raw):
    if not pd.isnull(subject_areas_raw):
        # convert to list with eval method
        subject_areas = eval(subject_areas_raw)
        return [s["area"] for s in subject_areas]
    else:
        return None


df.subject_areas = df.subject_areas.apply(convert_subject_areas_to_list)
```

```python
df["year"] = pd.DatetimeIndex(df.publication_date).year
```

## Publicações por professor

```python
publications_by_professor = df.groupby("professors").eid.count().reset_index()
publications_by_professor = publications_by_professor.sort_values(by="eid", ascending=False)
```

```python
publications_by_professor_top_10 = publications_by_professor.head(10).sort_values(by="eid", ascending=True)
```

```python
plt_1 = plt.figure(figsize=(14, 7))

plt.barh(publications_by_professor_top_10["professors"], publications_by_professor_top_10["eid"], color=color_blue)

plt.title('Professores com mais publicações')
plt.xlabel('Publicações')
plt.ylabel('Professor')

# Adicionando os números na frente das barras
for i, value in enumerate(publications_by_professor_top_10['eid']):
    plt.text(value + 3, i - 0.05, str(value), ha='center', va='bottom')

plt_1.show()
```

```python
## Pegando apenas dos últimos 10 anos
publications_by_professor_last_10_years = df[df.year >= 2014].groupby("professors").eid.count().reset_index()
publications_by_professor_last_10_years = publications_by_professor_last_10_years.sort_values(by="eid", ascending=False)
```

```python
publications_by_professor_top_10_last_10_years = publications_by_professor_last_10_years.head(10).sort_values(by="eid", ascending=True)
```

```python
plt_1 = plt.figure(figsize=(14, 7))

plt.barh(publications_by_professor_top_10_last_10_years["professors"], publications_by_professor_top_10_last_10_years["eid"], color=color_blue)

plt.title('Professores com mais publicações nos últimos 10 anos')
plt.xlabel('Publicações')
plt.ylabel('Professor')

# Adicionando os números na frente das barras
for i, value in enumerate(publications_by_professor_top_10_last_10_years['eid']):
    plt.text(value + 1, i - 0.05, str(value), ha='center', va='bottom')

plt_1.show()
```

## Citações por professor

```python
citations_by_professor = df.groupby("professors").citation_num.sum().reset_index()
citations_by_professor = citations_by_professor.sort_values(by="citation_num", ascending=False)
```

```python
citations_top_10 = citations_by_professor.head(10).sort_values(by="citation_num", ascending=True)
```

```python
plt_1 = plt.figure(figsize=(14, 7))

plt.barh(citations_top_10["professors"], citations_top_10["citation_num"], color=color_blue)

plt.title('Professores com mais citações')
plt.xlabel('Total de Citações')
plt.ylabel('Professor')

# Adicionando os números na frente das barras
for i, value in enumerate(citations_top_10['citation_num']):
    plt.text(value + 35, i - 0.05, str(int(value)), ha='center', va='bottom')

plt_1.show()
```

```python
citations_by_professor = df[df.year >= 2014].groupby("professors").citation_num.sum().reset_index()
citations_by_professor = citations_by_professor.sort_values(by="citation_num", ascending=False)
```

```python
citations_top_10 = citations_by_professor.head(10).sort_values(by="citation_num", ascending=True)
```

## Índice H

```python
df_h_index = df.groupby("professors").citation_num.apply(hindex).reset_index()
df_h_index.columns = ["professors", "h_index"]
df_h_index = df_h_index.sort_values(by="h_index", ascending=False)
```
