```python
df4 = []
for i in range(len(df3)):
    register = df3.iloc[i]
    subs = list(str(register['subject_areas'])[1:-2].split(', '))    
    for j in subs:
        line = {}
        line = register
        line['subject_areas'] = j.replace("'",'')
        df4.append(line)
```

```python
df4 = pd.DataFrame(df4).loc[:,['professors','subject_areas']]
len(df4.subject_areas.unique())
df5 =pd.get_dummies(data=df4,columns=['subject_areas'],prefix='',prefix_sep='')
```

```python
df5 = df5.groupby(by='professors',as_index=False).sum()
```
